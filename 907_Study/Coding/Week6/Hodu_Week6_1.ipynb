## 크레인 인형 뽑기
https://programmers.co.kr/learn/courses/30/lessons/64061

```python
def solution(board, moves):
    result = []
    answer = 0
    for i in moves:
        for j in range(len(board)):
            if board[j][i-1] != 0:
                result.append(board[j][i-1])
                board[j][i-1] = 0
                break
        if len(result) > 1:    
            if result[-2] == result[-1]:
                answer += 2
                del result[-1]
                del result[-1]
    return answer
```

## 모의고사
https://programmers.co.kr/learn/courses/30/lessons/42840

```python
def solution(answers):
    first = list(map(int, '12345' * (len(answers)//5 + 1)))
    second = list(map(int, '21232425' * (len(answers)//8 + 1)))
    third = list(map(int, '3311224455' * (len(answers)//10 +1)))
    
    f = 0
    s = 0
    t = 0
    
    for i in range(len(answers)):
        if answers[i] == first[i]:
            f += 1
        if answers[i] == second[i]:
            s += 1
        if answers[i] == third[i]:
            t += 1
    
    group = [f, s, t]
    answer = []
    
    for j in range(len(group)):
        if group[j] == group[j-1]:
            if group[j] >= group[j-2]:
                answer.append(j+1)
        elif group[j] > group[j-1]:
            if group[j] > group[j-2]:
                answer.clear()
                answer.append(j+1)
            elif group[j] == group[j-2]:
                answer.append(j+1)
                
    return answer
```

## 체육복
https://programmers.co.kr/learn/courses/30/lessons/42862

```python
def solution(n, lost, reserve):
    lost_n = lost.copy()
    for i in lost:
        if i in reserve:
            reserve.remove(i)
            lost_n.remove(i)
        elif i-1 in reserve and i-1 not in lost:
            reserve.remove(i-1)
            lost_n.remove(i)
        elif i+1 in reserve and i+1 not in lost:
            reserve.remove(i+1)
            lost_n.remove(i)
    answer = n - (len(lost_n))
    return answer
```

```python
#다른사람풀이
def strToInt(str):
    result = 0

    for idx, number in enumerate(str[::-1]):
        if number == '-':
            result *= -1
        else:
            result += int(number) * (10 ** idx)

    return result
```

```python
def solution(s, n):
    answer = s
    N = 0
    for i in s:
        if ord(i) >= 65 and ord(i) <=90:
            N += ord(i)
            N += n
            if N > 90:
                answer.replace(i, chr(N-26))
            else:
                answer.replace(i, chr(N))
        elif ord(i) >=97 and ord(i) <=122:
            N += ord(i)
            N += n
            if N > 122:
                answer.replace(i, chr(N-26))
            else:
                answer.replace(i, chr(N))
    return answer
```

```python
def solution(n):
    answer = sum(x for x in range(1, n+1) if n%x == 0)
    #리스트컴프리헨션이 sum안에 들어올 수 있음
    return answer
```
